```python
#print the fibonacci series of range m if n = 1 and print tribonacci series of range m if n = 2
n = int(input("Enter the value of n(1 or 2): "))
if n == 1:
  m = int(input("Enter the range of Fibonacci series: "))
  print("Fibonacci series:")
  a, b = 0, 1
  for i in range(m):
      print(a, end=" ")
      a, b = b, a + b
  print()
elif n == 2:
  m = int(input("Enter the range of Tribonacci series: "))
  print("Tribonacci series:")
  t0, t1, t2 = 0, 1, 1

  if m >= 1:
    print(t0, end=" ")
  if m >= 2:
    print(t1, end=" ")
  if m >= 3:
    print(t2, end=" ")

  for _ in range(3, m):
    tn = t0 + t1 + t2
    print(tn, end=" ")
    t0, t1, t2 = t1, t2, tn
  print() # Add a newline for better formatting after the series
else:
  print("Invalid input")
```

```python
#writw a python program to calculate the sum of digits until the single digit
num = int(input("Enter a number: "))
while num >= 10:
    sum_digits = 0
    while num > 0:
        sum_digits += num % 10
        num //= 10
    num = sum_digits

print("Single digit sum:", num)
```

```python
#write a python program to check whether the given no is palindrom or not
num = int(input("Enter a number: "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10

if num == rev:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
#write a python program to check whether the given no is armstrong or not
num = int(input("Enter a number: "))
temp = num
digits = len(str(num))
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** digits
    temp //= 10

if num == total:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
#write a program to find the vowels in the given string using function
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0

    for ch in s:
        if ch in vowels:
            count += 1

    return count

# Main program
string = input("Enter a string: ")
result = count_vowels(string)

print("Number of vowels in the string:", result)
```
